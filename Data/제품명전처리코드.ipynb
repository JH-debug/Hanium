```python
import pandas as pd
import re
```

```python
lee = pd.read_excel("의약품제품정보(제품명 전처리 전).xlsx")
```

```python
lee.head()
```

```python
lee["제품명"][0].split("(")[0]
```

```python
re.sub("[^ㄱ-ㅎ가-힣]","",lee["제품명"][0])
```

```python
def name_cut(x):
    x = x.split("(")[0]
    x = re.sub("[^ㄱ-ㅎ가-힣]","",x)
    return x                    
```

```python
result = name_cut(lee['제품명'][0]) ; result
```

```python
lee["제품명"] = lee["제품명"].apply(name_cut)
```

```python
lee.to_excel("의약품제품정보(지현 전처리 후).xlsx")
```
